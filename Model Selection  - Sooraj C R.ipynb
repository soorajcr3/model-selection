```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\DELL\Downloads\titanic_dataset  (1).csv")
```

```python
## Dropping the unwanted columns

df = df.drop('Name', axis =1)
df = df.drop('Ticket', axis =1)
df = df.drop('Fare', axis =1)
df = df.drop('Cabin', axis =1)
```

```python
# combining sibilins and parents
```

```python
df['Family'] = df['SibSp'] + df['Parch'] + 1

df = df.drop('SibSp', axis=1,)
df = df.drop('Parch', axis=1,)
```

```python
df.describe()
```

```python
## replacing age with median

df["Age"] =df["Age"].fillna(df["Age"].median())
```

```python
df.isnull().sum()
```

```python
#Replacing the NAN's with most frequently used one

print(df["Embarked"].mode())
df["Embarked"] = df["Embarked"].fillna("S")
```

```python
df.isnull().sum()
```

```python
## Replacing the categorical value into numerical value

df.Sex.replace(['male', 'female'], [1,0], inplace=True)
```

```python
df.head()
```

```python
## Replacing the categorical value into numerical value

df.Embarked.replace(['S', 'C', 'Q'], [1, 2, 3], inplace=True)
```

## KNN and SVM models

```python
X = df.drop('Survived',axis =1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state=42)
```

```python
## support vector classifier and
## KNN classifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC 

svc_classifier = SVC()
svc_classifier.fit(X_train,y_train)

from sklearn.model_selection import cross_val_score
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train,y_train)
```

```python
svc_predictions = svc_classifier.predict(X_test)
knn_predictions = knn_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
svc_accuracy = accuracy_score(y_test,svc_predictions)
knn_accuracy = accuracy_score(y_test,knn_predictions)
print("SVC accuracy :", svc_accuracy)
print("Knn accuracy :", knn_accuracy)
```

## K-fold

```python
from sklearn.model_selection import KFold
kfold_validator = KFold(10)
for train_index,test_index in kfold_validator.split(X,y):
    print('Training index :',train_index)
    print('Testing index :',test_index)
```

```python
from sklearn.model_selection import cross_val_score
cv_result = cross_val_score(knn_classifier,X,y,cv = kfold_validator)
cv_result
```

```python
np.mean(cv_result)
```

## stratified k-fold

```python
from sklearn.model_selection import StratifiedKFold
skfold_validator = StratifiedKFold(n_splits = 10)
for train_index,test_index in skfold_validator.split(X,y):
    print('Training index :',train_index)
    print('Testing index :',test_index)
```

```python
cv_result1 = cross_val_score(svc_classifier,X,y,cv = skfold_validator)
cv_result1
```

```python
np.mean(cv_result1)
```
